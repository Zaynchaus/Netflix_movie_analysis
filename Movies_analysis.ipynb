# project- netflix movies analysis

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python

netflix = pd.read_csv("netflix_movies.csv")
imdb = pd.read_csv("imdb_ratings.csv")
# netflix.head()
# imdb.head()
```

```python
#converting coloumns to numeric
netflix['duration'] = pd.to_numeric(netflix['duration'], errors='coerce')
imdb['imdb_rating'] = pd.to_numeric(imdb['imdb_rating'], errors='coerce')
imdb['imdb_votes'] = pd.to_numeric(imdb['imdb_votes'], errors='coerce')
```

```python
#merging datasets
df = pd.merge(netflix, imdb, on=["title", "release_year"], how="inner")

print("Merged dataset shape:", df.shape)
print(df.head())
```

```python
outdir = "analysis_outputs"
os.makedirs(outdir, exist_ok=True)

genre_counts = df['genre'].value_counts()
genre_counts.to_csv(f"{outdir}/genre_counts.csv")

avg_rating_genre = df.groupby('genre')['imdb_rating'].agg(['mean','count'])
avg_rating_genre.to_csv(f"{outdir}/avg_rating_by_genre.csv")

top20 = df.sort_values(by=['imdb_rating','imdb_votes'], ascending=[False, False]).head(20)
top20.to_csv(f"{outdir}/top_20_movies.csv", index=False)

director_stats = df.groupby('director')['imdb_rating'].agg(['mean','count'])
director_stats = director_stats[director_stats['count'] >= 3]
director_stats.to_csv(f"{outdir}/top_directors_stats.csv")

country_counts = df['country'].value_counts()
country_counts.to_csv(f"{outdir}/country_counts.csv")
```
